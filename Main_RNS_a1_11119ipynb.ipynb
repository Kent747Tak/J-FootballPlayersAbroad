```python
def extendLineXY (x1, x2, y1, y2):

    if x1<x2:
        x1b=x1-abs(x1-x2)
        x2b=x2+abs(x1-x2)
    else:
        x1b=x1+abs(x1-x2)
        x2b=x2-abs(x1-x2)
    if y1<y2:
        y1b=y1-abs(y1-y2)
        y2b=y2+abs(y1-y2)
    else:
        y1b=y1+abs(y1-y2)
        y2b=y2-abs(y1-y2)
    return (x1b, x2b, y1b, y2b)

#extendLineXY (x1, x2, y1, y2)
```

```python
### 間引いて読み込み　everySecondなど

def getFrames (file_dir, step_sec):

    from glob import glob
    import os, cv2
    #file_dir_out=file_dir+'/allFramesMoments'

    for file in glob (file_dir+'/*.mp4'):
        print (file)

    output_dir=file_dir+'/allFrames'

    cap = cv2.VideoCapture(file)

    ext='jpg'

    dir_path=file_dir+'/everySecond'
    os.makedirs(dir_path, exist_ok=True)
    base_path = os.path.join(dir_path, 'frame')

    print ('save_frame dir: ', dir_path)

    digit = len(str(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))))  #桁数

    fps = cap.get(cv2.CAP_PROP_FPS)
    fps_inv = 1 / fps

    sec = 0
    stop_sec=round(cap.get(cv2.CAP_PROP_FRAME_COUNT)*fps)

    while sec < stop_sec:
#        print (sec)
        n=round(fps*sec)

        cap.set(cv2.CAP_PROP_POS_FRAMES, n)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite('{}_{}_{:.2f}.{}'.format(base_path, str(n).zfill(digit), sec, ext), frame)

        else:
            break
        sec += step_sec

getFrames (file_dir, 0.1)
print ('finished')
```

```python
#YOLOする　　everySecondなどから間引いてYOLOできる

def YOLO_frames (file_dir):
    import cv2
    from ultralytics import YOLO
    from glob import glob
    import pandas as pd
    import os, re, math

    df_pose=pd.DataFrame()
    df_box=pd.DataFrame()

    os.makedirs(file_dir+'/results2', exist_ok=True)

    # モデルの読み込み。姿勢推論用のモデルデータを読み込む
    model = YOLO("yolov8n-pose.pt")

    for file in glob(file_dir+'/everySecond/*.jpg'):

        base = os.path.basename(file)
        #frame_0000_0.00.jpg
        sec =float(re.search('.*?_.*?_(.*?\..*?)\.jpg', base)[1])

        if sec % 0.5 ==0: # every 0.5 seconds

            frame = cv2.imread(file)
            result = model(frame, verbose=False)

            names = result[0].names
            classes = result[0].boxes.cls
            boxes = result[0].boxes

            for box, cls in zip(boxes, classes):

                x1, y1, x2, y2 =[0, 0, 0, 0]

                name = names[int(cls)]
                x1, y1, x2, y2 = [int(i) for i in box.xyxy[0]]

                print (x1, y1, x2, y2)
                df_box=pd.concat([df_box, pd.DataFrame([sec, name, x1, y1, x2, y2]).T], axis=0)

            if len(result[0].keypoints) == 0:
                continue

            # 姿勢分析結果のキーポイントを取得する
            keypoint = result[0].keypoints
            for i, point in enumerate(keypoint[0].data.tolist()[0]):
                df_pose=pd.concat([df_pose, pd.DataFrame([sec, i, point[0], point[1], point[2]]).T], axis=0)

            plottedFrame = result[0].plot()
            cv2.imwrite( file_dir+'//results2//'+base + "-resultPose.jpg", plottedFrame)


    df_pose.columns=['sec', 'point', 'x', 'y', 'possiblity']
    df_box.columns=['sec', 'item', 'xmin', 'ymin', 'xmax', 'ymax']

    df_pose=df_pose.sort_values(['sec', 'point'])
    df_box=df_box.sort_values('sec')

    excel_writer = pd.ExcelWriter(file_dir+'//'+'ResultData2.xlsx')
    df_pose.to_excel(excel_writer, 'pose', index=False)
    df_box.sort_values('sec').to_excel(excel_writer, 'box', index=False)

    excel_writer.close()
    return (df_pose, df_box)

(df_pose, df_box) = YOLO_frames (file_dir)
print ('finished!')
```
